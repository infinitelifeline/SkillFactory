```python
import json
from pprint import pprint
import pandas as pd

with open('..\\data\\recipe.json') as f:
    recipes = json.load(f)
    
    all_ingredients = set()

df = pd.DataFrame(recipes)

for i in df['ingredients']:
    for j in i:
        if j not in all_ingredients:
            all_ingredients.add(j)
            
len(all_ingredients)
```
