```python
stats = {}
count_post = 0
for record in response.json()['response']['items'][:]:
    title = record['text'][:30]
    if title:
        stats[title] = [record['comments']['count'],
        record['likes']['count'], record['reposts']['count'], record['date']]
        count_post += 1
    if count_post < 10:
        continue
    else:
        break

pprint(stats)
```
